```python
%matplotlib inline
from math import pi
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f(y):
    return (9000-y)/131.195335 # k
y = [i for i in range(0, 9010, 10)]
x = [] 
for i in y:
    x.append(f(i))

plt.plot(x, y, 'red', linewidth=2)
plt.scatter(68.6, 0, color='red')
plt.grid(True)
plt.title('n(M)')
plt.xlabel('М, мН*м')
plt.ylabel('n, об/мин.')
plt.text(66, 800, 'Mп', fontsize=12)


#plt.savefig('1.png', dpi=330)
```

```python
y1 = np.linspace(0, 7, 25)
x1 = [i*9.8 for i in y1]  # k = Iн / Mн

plt.plot(x1, y1, 'red', linewidth=2)
plt.scatter(0, 1.5)
plt.text(1, 1.4, 'Iн=1.5A')
plt.scatter(14.7, 0)
plt.text(13.9, 0.4, 'Mн=14.7мН*м')
plt.scatter(14.7, 1.5, color='red')
plt.grid(True)
plt.title('Зависимость потребляемого тока от момента')
plt.xlabel('М, мН*м')
plt.ylabel('I, A')

#plt.savefig('2.png', dpi=300)
```

```python
'Мощность найдём как I*U в номинальном режиме и посчитаем к, P1 = 40.5'


p1 = [m*(40.5/14.7) for m in x]
p2 = [(i[0]*i[1]*pi)/(30*1000) for i in zip(x,y)]

plt.figure(1)
plt.plot(x, p1, 'red', linewidth=2, label='P1 - потребляемая')
plt.plot(x, p2, 'b', linewidth=2, label='P2 - на валу')
plt.scatter(14.7, 0)
plt.scatter(0, 40.5)
plt.scatter(14.7, 40.5, color='red')
plt.scatter(x[len(x)//2], p2[len(p2)//2], color='b')
plt.grid(True)
plt.title('Зависимость потребляемой мощности \nи мощности на валу от момента')
plt.xlabel('М, мН*м')
plt.ylabel('P, Вт')
plt.legend(loc=0, borderaxespad=1)
plt.text(10, 50, 'Pн1', fontsize=12)
plt.text(12, 18, 'Мн', fontsize=12)
plt.text(26, 22, 'P2max=16', fontsize=10)
print(p1[::-100])
print(p2[::100])
#plt.savefig('3.png', dpi=330)
```
